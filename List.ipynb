### Looping through List

```python
players = ['Messi','Ronaldo','Neymar','Haaland','Mbappe']
```

```python
for player in players:
    print(player)
```

```python
for i in range(1,5):
    print(players[i])
```

### List Comprehension

```python
clubs = ["JY Club","Manang Marsyandi","3 stars Club","ACHS Club","RCT","Sathdobato"]
```

```python
club_clubs=[] # clubs name containing Club in their name ex: JY Club
for club in clubs:
    if "Club" in club:
        club_clubs.append(club)
```

```python
print(club_clubs)
```

```python
# Another way of doing it
club_with_club_in_their_name=[club for club in clubs if "Club" in club]  
```

```python
club_with_club_in_their_name
```

```python
new_list=[x for x in range(6)]
new_list
```
