### passing value in a function

```python
def greet_user(user):
    print("Hello ", user)
```

```python
greet_user("Admin")
```

#### passing multiple values in function

```python
def multiply(x, y):
    return x*y
```

```python
multiply(4,5)
```

```python
def power(num1, num2):
    return f'{num1} power {num2} is : {num1**num2}'
```

```python
power(2,5)
```

#### function with default args value

```python
def power(num1, num2=2):
    return num2
```
