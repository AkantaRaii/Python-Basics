#Set operations

```python
novels1 = {"Catch 22","Scicilian","Godfather","The Last Don"}
novels2 = {"1Q84","IT","Godfather","Scicilian","Fools Die"}
```

```python
novels3 = novels1.union(novels2)
novels3
```

```python
novels4 = novels1.intersection(novels2)
novels4
```

```python
novels5 = novels1.difference(novels2)
novels5
```

```python
novels5 = novels1.symmetric_difference(novels2)
novels5
```

```python
novels1 & novels2
```

```python
novels1 | novels2
```

```python
novels1 - novels2
```

```python
novels1 ^ novels2
```
