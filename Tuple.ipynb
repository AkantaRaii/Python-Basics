### Unpacking a tuple

```python
numbers=(2,3,4)
```

```python
two,three,four=numbers
print(two)
print(three)
print(four)
```

```python
players=("Messi","Ronaldo","Neymar","Benzema")
```

```python
player1,player2,player3,player4=players
print(player1)
print(player2)
print(player3)
print(player4)
```

### joining tuples

```python
odd=(1,2,5,)
even=(2,4,5)
numbers=odd+even
print(type(numbers))
print(numbers)
```
